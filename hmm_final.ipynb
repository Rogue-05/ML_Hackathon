```python
import os, re, zipfile, json, math, random, string, collections, statistics, pathlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import VarianceThreshold
from typing import List, Dict, Set, Iterable, Tuple
import pickle
from tqdm import tqdm
import time

SEED = 1337
random.seed(SEED)
np.random.seed(SEED)
plt.rcParams["figure.figsize"] = (8, 4)
ALPH = "abcdefghijklmnopqrstuvwxyz"
ALPH_SET = set(ALPH)

try:
    import google.colab
    IN_COLAB = True
except Exception:
    IN_COLAB = False

import os, pathlib

if IN_COLAB:
    BASE = "/content/"
    DATA_ROOT = f"{BASE}/drive/MyDrive/Colab Notebooks/data"
    OUT_ROOT  = f"{BASE}/outputs"
    os.makedirs(DATA_ROOT, exist_ok=True)
    os.makedirs(OUT_ROOT, exist_ok=True)
else:
    try:
        NB_DIR = pathlib.Path(__file__).parent
    except NameError:
        NB_DIR = pathlib.Path(os.getcwd())

    data_candidates = [
        NB_DIR / "data",
        NB_DIR / "Data",
        NB_DIR.parent / "data",
        NB_DIR.parent / "Data"
    ]

    DATA_ROOT = None
    for candidate in data_candidates:
        if candidate.exists() and (candidate / "corpus.txt").exists():
            DATA_ROOT = str(candidate.resolve())
            break

    if DATA_ROOT is None:
        raise FileNotFoundError("Could not find data directory with corpus.txt")

    OUT_ROOT = str((NB_DIR / "outputs").resolve())
    os.makedirs(OUT_ROOT, exist_ok=True)
    print(f"Running locally. DATA_ROOT={DATA_ROOT}")

corpus_path = os.path.join(DATA_ROOT, "corpus.txt")
test_path = os.path.join(DATA_ROOT, "test.txt")

if not os.path.exists(corpus_path):
    raise FileNotFoundError(f"Missing corpus file: {corpus_path}")
if not os.path.exists(test_path):
    raise FileNotFoundError(f"Missing test file: {test_path}")

print(f"Found corpus file: {corpus_path}")
print(f"Found test file: {test_path}")

def normalize_word(w: str) -> str:
    w = w.strip().lower()
    w = re.sub(r'[^a-z]', '', w)
    return w

def load_words(path: str, min_len: int = 2, max_len: int | None = None):
    words = []
    with open(path, 'r', encoding='utf-8', errors='ignore') as f:
        for line in f:
            w = normalize_word(line)
            if not w:
                continue
            if len(w) < min_len:
                continue
            if max_len is not None and len(w) > max_len:
                continue
            if all(c in ALPH for c in w):
                words.append(w)
    return words

def dedupe(seq):
    seen = set(); out = []
    for x in seq:
        if x not in seen:
            out.append(x); seen.add(x)
    return out

class HMMOracle:
    def __init__(self):
        self.models: Dict[int, Dict[str, np.ndarray]] = {}
        self.alphabet = 'abcdefghijklmnopqrstuvwxyz'
        self.char_to_index = {c: i for i, c in enumerate(self.alphabet)}
        self.index_to_char = {i: c for i, c in enumerate(self.alphabet)}
        self.trained_lengths: Set[int] = set()

    def _normalize_words(self, words: Iterable[str]) -> List[str]:
        out = []
        for w in words:
            w = (w or "").strip().lower()
            if w and all(c in self.alphabet for c in w):
                out.append(w)
        return out

    def train_from_words(self, words: Iterable[str]) -> None:
        words = self._normalize_words(words)
        words_by_length: Dict[int, List[str]] = defaultdict(list)
        for w in words:
            words_by_length[len(w)].append(w)

        self.models.clear()
        self.trained_lengths.clear()

        for length, word_list in words_by_length.items():
            if length < 2:
                continue

            pi_counts = np.ones(26, dtype=np.float64)
            transition_counts = np.ones((length - 1, 26, 26), dtype=np.float64)

            for w in word_list:
                pi_counts[self.char_to_index[w[0]]] += 1.0
                for t in range(length - 1):
                    i = self.char_to_index[w[t]]
                    j = self.char_to_index[w[t + 1]]
                    transition_counts[t, i, j] += 1.0

            pi = pi_counts / np.sum(pi_counts)
            A = np.zeros_like(transition_counts)
            for t in range(length - 1):
                row_sums = transition_counts[t].sum(axis=1, keepdims=True)
                A[t] = transition_counts[t] / np.maximum(row_sums, 1e-12)

            self.models[length] = {"pi": pi, "A": A}
            self.trained_lengths.add(length)

    def get_letter_probabilities(self, masked_word: str, guessed_letters: Set[str]) -> np.ndarray:
        masked_word = (masked_word or "").strip().lower()
        guessed_letters = {c.lower() for c in (guessed_letters or set()) if c.isalpha()}
        L = len(masked_word)

        if L not in self.models:
            remain = [i for i in range(26) if self.index_to_char[i] not in guessed_letters]
            out = np.zeros(26, dtype=np.float64)
            if remain:
                out[remain] = 1.0 / len(remain)
            return out

        model = self.models[L]
        pi, A = model["pi"], model["A"]

        remaining_indices = {self.char_to_index[c] for c in self.alphabet if c not in guessed_letters}
        if not remaining_indices:
            return np.zeros(26, dtype=np.float64)

        alpha = np.zeros((L, 26), dtype=np.float64)
        for j in range(26):
            if masked_word[0] == self.index_to_char[j]:
                alpha[0, j] = pi[j]
            elif masked_word[0] == '_' and j in remaining_indices:
                alpha[0, j] = pi[j]

        for t in range(1, L):
            prev = alpha[t - 1]
            for j in range(26):
                s = np.dot(prev, A[t - 1, :, j])
                if masked_word[t] == self.index_to_char[j]:
                    alpha[t, j] = s
                elif masked_word[t] == '_' and j in remaining_indices:
                    alpha[t, j] = s

        beta = np.zeros((L, 26), dtype=np.float64)
        beta[L - 1, :] = 1.0
        for t in range(L - 2, -1, -1):
            for j in range(26):
                total = 0.0
                for k in range(26):
                    if masked_word[t + 1] == self.index_to_char[k] or (masked_word[t + 1] == '_' and k in remaining_indices):
                        total += beta[t + 1, k] * A[t, j, k]
                beta[t, j] = total

        final = np.zeros(26, dtype=np.float64)
        Z = np.sum(alpha[L - 1])
        if Z <= 0:
            for j in remaining_indices:
                final[j] = 1.0 / max(1, len(remaining_indices))
            return final

        for t in range(L):
            if masked_word[t] == '_':
                for j in remaining_indices:
                    final[j] += (alpha[t, j] * beta[t, j]) / Z

        s = final.sum()
        if s > 0:
            final /= s
        return final

    def suggest_next_letters(self, masked_word: str, guessed_letters: Set[str], top_k: int = 10):
        probs = self.get_letter_probabilities(masked_word, guessed_letters)
        candidates = [(self.index_to_char[i], float(probs[i])) for i in range(26) if probs[i] > 0]
        candidates.sort(key=lambda x: x[1], reverse=True)
        return candidates[:top_k]

from typing import Optional

def _hmmoracle_train(self, corpus_file: Optional[str] = None):
    def _load_words_from(path: str):
        try:
            with open(path, 'r', encoding='utf-8') as f:
                return [line.strip().lower() for line in f if line.strip()]
        except Exception as e:
            raise FileNotFoundError(f"Could not read corpus file: {path} ({e})")

    if corpus_file is not None:
        words = _load_words_from(corpus_file)
    else:
        words = globals().get('corpus', None)
        if words is None:
            try:
                 words = load_words(os.path.join(DATA_ROOT, "corpus.txt"))
            except FileNotFoundError:
                 raise ValueError("Global 'corpus' not found and corpus.txt not found in DATA_ROOT. Provide corpus_file or run data loading earlier.")

    words = [''.join(ch for ch in w if ch in ALPH_SET) for w in words]
    if hasattr(self, 'train_from_words'):
        self.train_from_words(words)
    elif hasattr(self, 'train'):
        return self.train(words)
    else:
        raise AttributeError("HMMOracle has no training method available.")

if 'HMMOracle' in globals():
    HMMOracle.train = _hmmoracle_train
    print("HMMOracle.train shim installed.")
else:
    print("Warning: HMMOracle not defined yet; run the HMM class cell first, then re-run this cell.")


class HangmanEnvironment:
    def __init__(self, word_list, hmm_oracle, max_lives=6):
        self.word_list = word_list
        self.hmm_oracle = hmm_oracle
        self.max_lives = max_lives
        self.alphabet = 'abcdefghijklmnopqrstuvwxyz'
        self.char_to_index = {c: i for i, c in enumerate(self.alphabet)}
        self.reset()

    def reset(self):
        self.target_word = random.choice(self.word_list).lower()
        self.masked_word = ['_'] * len(self.target_word)
        self.guessed_letters = set()
        self.lives = self.max_lives
        self.wrong_guesses = 0
        self.repeated_guesses = 0
        self.game_over = False
        self.won = False
        self.hmm_probs = self._get_hmm_probabilities()
        return self._get_state()

    def _get_hmm_probabilities(self):
        masked_string = ''.join(self.masked_word)
        probs = self.hmm_oracle.get_letter_probabilities(masked_string, self.guessed_letters)
        return probs

    def _get_state(self):
        lives = self.lives

        masked_representation = tuple(
            self.char_to_index.get(char, -1) for char in self.masked_word
        )

        guessed_representation = tuple(
            (self.alphabet[i] in self.guessed_letters) for i in range(26)
        )
        probs_copy = self.hmm_probs.copy()
        for i in range(26):
            if self.alphabet[i] in self.guessed_letters:
                probs_copy[i] = -1.0

        top_hmm_indices = np.argsort(probs_copy)[::-1]
        valid_top_hmm_indices = [
            idx for idx in top_hmm_indices if probs_copy[idx] >= 0
        ][:3]

        while len(valid_top_hmm_indices) < 3:
            valid_top_hmm_indices.append(-1)

        hmm_top_guesses = tuple(valid_top_hmm_indices)

        state = (lives, masked_representation, guessed_representation, hmm_top_guesses)

        return state


    def _get_state_vector(self):
        return {
            'masked_word': ''.join(self.masked_word),
            'guessed': self.guessed_letters.copy(),
            'lives': self.lives,
            'hmm_probs': self.hmm_probs.copy()
        }


    def step(self, action):
        letter = self.alphabet[action]
        if letter in self.guessed_letters:
            self.repeated_guesses += 1
            reward = -2.0
            return self._get_state(), reward, self.game_over, {
                'repeated': True,
                'won': self.won,
                'letter': letter
            }
        self.guessed_letters.add(letter)

        if letter in self.target_word:
            reward = 5
            for i, char in enumerate(self.target_word):
                if char == letter:
                    self.masked_word[i] = letter

            if '_' not in self.masked_word:
                self.game_over = True
                self.won = True
                reward = 100
        else:
            self.lives -= 1
            self.wrong_guesses += 1
            reward = -5

            if self.lives <= 0:
                self.game_over = True
                self.won = False
                reward = -100.0
        if not self.game_over:
            self.hmm_probs = self._get_hmm_probabilities()
        next_state = self._get_state()
        info = {
            'repeated': False,
            'won': self.won,
            'correct': letter in self.target_word,
            'letter': letter
        }

        return next_state, reward, self.game_over, info

    def get_valid_actions(self):
        """Get list of valid action indices (not yet guessed)."""
        return [i for i in range(26) if self.alphabet[i] not in self.guessed_letters]


class QLearningAgent:
    """
    Tabular Q-Learning agent for Hangman.
    Uses state discretization to make Q-table manageable.
    """
    def __init__(self, action_size=26, learning_rate=0.1, discount_factor=0.95):
        """
        Args:
            action_size: Number of possible actions (26 letters)
            learning_rate: Learning rate (alpha)
            discount_factor: Discount factor (gamma)
        """
        self.action_size = action_size
        self.alpha = learning_rate
        self.gamma = discount_factor
        self.epsilon = 1.0
        self.epsilon_min = 0.05
        self.epsilon_decay = 0.995
        self.q_table = defaultdict(lambda: np.zeros(action_size))
        self.training_step = 0
        self.alphabet = 'abcdefghijklmnopqrstuvwxyz'

    def get_q_value(self, state, action):
        return self.q_table[state][action]

    def select_action(self, state, valid_actions=None):
        if valid_actions is None:
            valid_actions = list(range(self.action_size))

        if random.random() < self.epsilon:
            return random.choice(valid_actions)
        q_values = self.q_table[state]
        valid_q_values = [(action, q_values[action]) for action in valid_actions]
        if all(q == 0 for _, q in valid_q_values):
            return random.choice(valid_actions)
        best_action = max(valid_q_values, key=lambda x: x[1])[0]
        return best_action


    def select_action_with_hmm(self, state, env, exploration_weight=0.5):
        valid_actions = env.get_valid_actions()
        if not valid_actions:
            return random.randint(0, self.action_size - 1)

        if random.random() < self.epsilon:
            return random.choice(valid_actions)

        q_values = self.q_table[state]
        hmm_probs = env.hmm_probs

        valid_q = np.array([q_values[a] for a in valid_actions])
        valid_hmm = np.array([hmm_probs[a] for a in valid_actions])

        min_q = valid_q.min()
        if min_q < 0:
            normalized_q = valid_q - min_q
        else:
            normalized_q = valid_q

        hmm_sum = valid_hmm.sum()
        if hmm_sum > 0:
            normalized_hmm = valid_hmm / hmm_sum
        else:
            normalized_hmm = np.ones_like(valid_hmm) / len(valid_hmm)

        combined_scores = (normalized_q + 1e-6) * ((normalized_hmm ** exploration_weight) + 1e-6)


        best_idx = np.argmax(combined_scores)
        return valid_actions[best_idx]


    def update(self, state, action, reward, next_state, done, next_valid_actions=None):
        current_q = self.q_table[state][action]

        if done:
            target_q = reward
        else:
            if next_valid_actions:
                next_q_values = [self.q_table[next_state][a] for a in next_valid_actions]
                max_next_q = max(next_q_values) if next_q_values else 0
            else:
                max_next_q = 0


            target_q = reward + self.gamma * max_next_q
        self.q_table[state][action] = current_q + self.alpha * (target_q - current_q)

        self.training_step += 1

    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

    def save(self, filepath):
        data = {
            'q_table': dict(self.q_table),
            'epsilon': self.epsilon,
            'alpha': self.alpha,
            'gamma': self.gamma,
            'epsilon_decay': self.epsilon_decay
        }
        with open(filepath, 'wb') as f:
            pickle.dump(data, f)
        print(f"Model saved to {filepath}")

    def load(self, filepath):
        with open(filepath, 'rb') as f:
            data = pickle.load(f)
        self.q_table = defaultdict(lambda: np.zeros(self.action_size), data['q_table'])
        self.epsilon = data.get('epsilon', 1.0)
        self.alpha = data.get('alpha', 0.1)
        self.gamma = data.get('gamma', 0.95)
        self.epsilon_decay = data.get('epsilon_decay', 0.995)
        print(f"Model loaded from {filepath}")


def train_agent(agent, env, num_episodes=1000, eval_every=50, use_hmm_guidance=True):
    episode_rewards = []
    episode_wins = []
    episode_wrong_guesses = []
    episode_repeated_guesses = []
    q_table_sizes = []

    print("Starting Q-Learning training...")
    print(f"Use HMM Guidance: {use_hmm_guidance}")

    for episode in tqdm(range(num_episodes)):
        state = env.reset()
        episode_reward = 0

        while True:
            if use_hmm_guidance:
                action = agent.select_action_with_hmm(state, env, exploration_weight=0.5)
            else:
                valid_actions = env.get_valid_actions()
                action = agent.select_action(state, valid_actions)

            next_state, reward, done, info = env.step(action)

            next_valid_actions = env.get_valid_actions() if not done else None

            agent.update(state, action, reward, next_state, done, next_valid_actions)

            episode_reward += reward
            state = next_state

            if done:
                break

        agent.decay_epsilon()

        episode_rewards.append(episode_reward)
        episode_wins.append(1 if env.won else 0)
        episode_wrong_guesses.append(env.wrong_guesses)
        episode_repeated_guesses.append(env.repeated_guesses)
        q_table_sizes.append(len(agent.q_table))

        if (episode + 1) % eval_every == 0:
            recent_wins = np.mean(episode_wins[-eval_every:])
            recent_reward = np.mean(episode_rewards[-eval_every:])
            recent_wrong = np.mean(episode_wrong_guesses[-eval_every:])
            recent_repeated = np.mean(episode_repeated_guesses[-eval_every:])

            print(f"\nEpisode {episode + 1}/{num_episodes}")
            print(f"  Win Rate: {recent_wins:.2%}")
            print(f"  Avg Reward: {recent_reward:.2f}")
            print(f"  Avg Wrong: {recent_wrong:.2f}")
            print(f"  Avg Repeated: {recent_repeated:.2f}")
            print(f"  Epsilon: {agent.epsilon:.3f}")
            print(f"  Q-Table Size: {len(agent.q_table)}")

    return {
        'rewards': episode_rewards,
        'wins': episode_wins,
        'wrong_guesses': episode_wrong_guesses,
        'repeated_guesses': episode_repeated_guesses,
        'q_table_sizes': q_table_sizes
    }

def evaluate_agent(agent, env, num_games=2000, use_hmm_guidance=True):
    original_epsilon = agent.epsilon
    agent.epsilon = 0.0
    wins = 0
    total_wrong = 0
    total_repeated = 0
    game_details = []

    print(f"\nEvaluating on {num_games} games...")

    for game in tqdm(range(num_games)):
        state = env.reset()
        game_wrong = 0
        game_repeated = 0
        while True:

            if use_hmm_guidance:
                action = agent.select_action_with_hmm(state, env, exploration_weight=0.5)
            else:
                valid_actions = env.get_valid_actions()
                action = agent.select_action(state, valid_actions)
            next_state, reward, done, info = env.step(action)
            if info['repeated']:
                game_repeated += 1

            state = next_state

            if done:
                if env.won:
                    wins += 1
                game_wrong = env.wrong_guesses
                game_repeated = env.repeated_guesses
                break
        total_wrong += game_wrong
        total_repeated += game_repeated
        game_details.append({
            'word': env.target_word,
            'won': env.won,
            'wrong': game_wrong,
            'repeated': game_repeated
        })
    agent.epsilon = original_epsilon
    success_rate = wins / num_games
    final_score = (success_rate * 2000) - (total_wrong * 5) - (total_repeated * 2)
    results = {
        'success_rate': success_rate,
        'wins': wins,
        'total_games': num_games,
        'total_wrong': total_wrong,
        'total_repeated': total_repeated,
        'avg_wrong': total_wrong / num_games,
        'avg_repeated': total_repeated / num_games,
        'final_score': final_score,
        'game_details': game_details
    }

    print("EVALUATION RESULTS")
    print(f"Success Rate: {success_rate:.2%} ({wins}/{num_games})")
    print(f"Total Wrong Guesses: {total_wrong}")
    print(f"Total Repeated Guesses: {total_repeated}")
    print(f"Avg Wrong per Game: {results['avg_wrong']:.2f}")
    print(f"Avg Repeated per Game: {results['avg_repeated']:.2f}")
    print(f"\nFINAL SCORE: {final_score:.2f}")
    return results


def plot_training_metrics(metrics, save_path='training_plots.png'):
    fig, axes = plt.subplots(1, 2, figsize=(15, 10))

    axes[0].plot(metrics['rewards'], alpha=0.3, label='Episode Reward')
    if len(metrics['rewards']) >= 50:
        axes[0].plot(np.convolve(metrics['rewards'], np.ones(50)/50, mode='valid'),
                       label='50-episode MA', linewidth=2)
    axes[0].set_title('Episode Rewards', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Episode')
    axes[0].set_ylabel('Total Reward')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)


    axes[1].plot(metrics['wrong_guesses'], alpha=0.3, label='Wrong Guesses')
    if len(metrics['wrong_guesses']) >= 50:
        axes[1].plot(np.convolve(metrics['wrong_guesses'], np.ones(50)/50, mode='valid'),
                       label='50-episode MA', linewidth=2)
    axes[1].set_title('Wrong Guesses per Episode', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Episode')
    axes[1].set_ylabel('Wrong Guesses')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    print(f"\nTraining plots saved to {save_path}")
    plt.show()


def demo_game(agent, env, num_games=5, use_hmm_guidance=True):
    original_epsilon = agent.epsilon
    agent.epsilon = 0.0

    print("DEMO GAMEPLAY")

    for game_num in range(num_games):
        state = env.reset()

        print(f"\nGame {game_num + 1}")
        print(f"Target Word: {env.target_word} (Length: {len(env.target_word)})")

        turn = 1
        while True:
            state_info = env._get_state_vector()
            masked = state_info['masked_word']
            guessed = sorted(list(state_info['guessed']))
            lives = state_info['lives']

            print(f"\nTurn {turn}:")
            print(f"  Current: {masked}")
            print(f"  Guessed: {guessed if guessed else 'None'}")
            print(f"  Lives: {lives}/{env.max_lives}")
            hmm_probs = state_info['hmm_probs']
            valid_actions = env.get_valid_actions()
            valid_hmm = [(env.alphabet[i], hmm_probs[i]) for i in valid_actions]
            valid_hmm.sort(key=lambda x: x[1], reverse=True)
            top_hmm = valid_hmm[:3]
            print(f"  HMM Suggestions: {', '.join([f'{c}({p:.3f})' for c, p in top_hmm])}")
            if use_hmm_guidance:
                action = agent.select_action_with_hmm(state, env, exploration_weight=0.5)
            else:
                action = agent.select_action(state, valid_actions)

            letter = env.alphabet[action]

            print(f"  Agent guesses: {letter.upper()}")

            next_state, reward, done, info = env.step(action)

            if info['repeated']:
                print(f"REPEATED! Penalty: {reward}")
            elif info['correct']:
                print(f"Correct! Reward: {reward}")
            else:
                print(f"Wrong! Penalty: {reward}")

            state = next_state
            turn += 1

            if done:
                if env.won:
                    print(f"WON! Word was: {env.target_word.upper()}")
                else:
                    print(f"LOST! Word was: {env.target_word.upper()}")
                print(f"Turns: {turn - 1}")
                print(f"Wrong guesses: {env.wrong_guesses}")
                print(f"Repeated guesses: {env.repeated_guesses}")
                break
    agent.epsilon = original_epsilon


if __name__ == "__main__":
    print("HANGMAN Q-LEARNING AGENT WITH HMM ORACLE (Optimized)")

    print("\n[1/6] Loading word corpus...")
    try:
        with open(os.path.join(DATA_ROOT, 'corpus.txt'), 'r', encoding='utf-8') as f:
            train_words = [line.strip().lower() for line in f if line.strip()]
        print(f"Loaded {len(train_words)} words from corpus.txt")
    except FileNotFoundError:
        print("Error: corpus.txt not found. Using sample words for demo.")
        train_words = ["machine", "learning", "python", "algorithm", "neural",
                       "network", "training", "testing", "model", "data"] * 100
    try:
        with open(os.path.join(DATA_ROOT, 'test.txt'), 'r', encoding='utf-8') as f:
            test_words = [line.strip().lower() for line in f if line.strip()]
        print(f"Loaded {len(test_words)} words from test.txt")
    except FileNotFoundError:
        print("Warning: test.txt not found. Using 20% of corpus for testing.")
        random.shuffle(train_words)
        split_idx = int(0.8 * len(train_words))
        test_words = train_words[split_idx:]
        train_words = train_words[:split_idx]

    print("\n[2/6] Training HMM Oracle...")
    hmm_oracle = HMMOracle()
    try:
        hmm_oracle.train(os.path.join(DATA_ROOT, 'corpus.txt'))
    except:
        print("Note: Using default HMM training path")
        hmm_oracle.train()

    print("\n[3/6] Data summary...")
    print(f"Training words: {len(train_words)}")
    print(f"Testing words: {len(test_words)}")

    print("\n[4/6] Creating environments...")
    train_env = HangmanEnvironment(train_words, hmm_oracle)
    test_env = HangmanEnvironment(test_words, hmm_oracle)

    print("\n[5/6] Creating and training Q-Learning agent...")
    agent = QLearningAgent(
        action_size=26,
        learning_rate=0.05,
        discount_factor=0.99
    )
    agent.epsilon_decay = 0.998


    start_time = time.time()
    metrics = train_agent(
        agent,
        train_env,
        num_episodes=3000,
        eval_every=200,
        use_hmm_guidance=True
    )
    training_time = time.time() - start_time

    print(f"\nTraining completed in {training_time:.2f} seconds")
    print(f"Final Q-Table size: {len(agent.q_table)} states")

    plot_training_metrics(metrics, os.path.join(OUT_ROOT, 'q_learning_training_plots_optimized.png'))

    model_path = os.path.join(OUT_ROOT, 'hmm_model.pkl')
    agent.save(model_path)
    print(f"Model saved to {model_path}")

    print("\n[6/6] Evaluating on test set...")
    results_optimized = evaluate_agent(
        agent,
        test_env,
        num_games=len(test_words),
        use_hmm_guidance=True
    )

    print("FINAL SUMMARY (Optimized Agent)")
    print(f"Training Episodes: {len(metrics['rewards'])}")
    print(f"Training Time: {training_time:.2f}s")
    print(f"Q-Table Size: {len(agent.q_table)} states")
    print(f"Final Success Rate: {results_optimized['success_rate']:.2%}")
    print(f"Final Score: {results_optimized['final_score']:.2f}")

    initial_win_rate = 0.1770
    accuracy_increase = results_optimized['success_rate'] - initial_win_rate
    print(f"\nAccuracy Increase compared to initial run ({initial_win_rate:.2%}): {accuracy_increase:.2%}")
```
